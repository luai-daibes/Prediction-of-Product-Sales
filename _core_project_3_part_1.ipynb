```python
import glob
import pandas as pd
folder = "/content/drive/MyDrive/AXSOSACADEMY/06-AdvancedML/Week24/"
crime_files = sorted(glob.glob(folder+"*.csv"))
df = pd.concat([pd.read_csv(f) for f in crime_files])
```

```python
import os
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y %I:%M:%S %p')
df = df.set_index('Date')
display(df.index)
display(df.info())
```

```python
yearly_crime_count = df.resample('Y').size()
print("Total crimes per year:")
display(yearly_crime_count)
crime_type_yearly_count = df.groupby('Primary Type').resample('Y').size()
print("\nCrimes per year by Primary Type:")
display(crime_type_yearly_count)
```

```python
am_rush_start = '06:00:00'
am_rush_end = '09:00:00'
pm_rush_start = '16:00:00'
pm_rush_end = '19:00:00'
df['Time'] = df.index.time
am_rush_df = df[(df['Time'] >= pd.to_datetime(am_rush_start).time()) & (df['Time'] <= pd.to_datetime(am_rush_end).time())]
pm_rush_df = df[(df['Time'] >= pd.to_datetime(pm_rush_start).time()) & (df['Time'] <= pd.to_datetime(pm_rush_end).time())]
am_rush_count = len(am_rush_df)
pm_rush_count = len(pm_rush_df)
print(f"Number of crimes during AM rush hour ({am_rush_start}-{am_rush_end}): {am_rush_count}")
print(f"Number of crimes during PM rush hour ({pm_rush_start}-{pm_rush_end}): {pm_rush_count}")
am_rush_top_crimes = am_rush_df['Primary Type'].value_counts().head(5)
pm_rush_top_crimes = pm_rush_df['Primary Type'].value_counts().head(5)
print("\nTop 5 most common crimes during AM rush hour:")
display(am_rush_top_crimes)
print("\nTop 5 most common crimes during PM rush hour:")
display(pm_rush_top_crimes)
motor_vehicle_theft_am = am_rush_df[am_rush_df['Primary Type'] == 'MOTOR VEHICLE THEFT'].shape[0]
motor_vehicle_theft_pm = pm_rush_df[pm_rush_df['Primary Type'] == 'MOTOR VEHICLE THEFT'].shape[0]
print(f"\nMotor Vehicle Thefts during AM rush hour: {motor_vehicle_theft_am}")
print(f"Motor Vehicle Thefts during PM rush hour: {motor_vehicle_theft_pm}")
```

```python
monthly_crime_count = df.groupby(df.index.month).size()
# lets change month from number to name
monthly_crime_count.index = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
print("Total crimes per month:")
display(monthly_crime_count)
crime_type_monthly_count = df.groupby(['Primary Type', df.index.month]).size()
print("\nCrimes per month by Primary Type:")
display(crime_type_monthly_count)
```

```python
print("Crime Analysis Summary")
print("----------------------")
print("\nYearly Crime Trends:")
print(f"The total number of crimes has generally decreased across the years. The yearly crime counts are as follows:")
display(yearly_crime_count)
print("\nRush Hour Crime Analysis:")
print(f"There were {am_rush_count} crimes during AM rush hour (6 AM - 9 AM) and {pm_rush_count} crimes during PM rush hour (4 PM - 7 PM).")
print("Crimes are more common during PM rush hour.")
print("\nTop 5 most common crimes during AM rush hour:")
display(am_rush_top_crimes)
print("\nTop 5 most common crimes during PM rush hour:")
display(pm_rush_top_crimes)
print(f"\nMotor Vehicle Thefts during AM rush hour: {motor_vehicle_theft_am}")
print(f"Motor Vehicle Thefts during PM rush hour: {motor_vehicle_theft_pm}")
print("Motor Vehicle Thefts are more common during PM rush hour.")
print("\nMonthly Crime Patterns:")
print("The months with the most and least crime overall are:")
display(monthly_crime_count)
print("Overall, summer months (July and August) tend to have higher crime rates, while winter months (February and December) have lower rates.")
print("\nMonthly crime counts by Primary Type:")
display(crime_type_monthly_count)
print("Further analysis of 'Crimes per month by Primary Type' would reveal if any individual crime types do not follow the overall monthly pattern.")
```
